# Bank Management System

```python
import pickle
import os

# Load accounts
if os.path.exists("accounts.pkl"):
    with open("accounts.pkl", "rb") as f:
        accounts = pickle.load(f)
else:
    accounts = {}

# Load last account number
if os.path.exists("meta.pkl"):
    with open("meta.pkl", "rb") as f:
        account_number = pickle.load(f)
else:
    account_number = 1000

# Load transaction history
if os.path.exists("transactions.pkl"):
    with open("transactions.pkl", "rb") as f:
        transactions = pickle.load(f)
else:
    transactions = []
    
print("="*45)
print("   WELCOME TO BANK MANAGEMENT SYSTEM   ")
print("="*45)


while True:
    print("\n------ BANK MENU ------")
    print("1. Open New Account")
    print("2. View Account Details")
    print("3. Deposit Money")
    print("4. Withdraw Money")
    print("5. Transfer Money")
    print("6. View Transaction History")
    print("7. Exit")


    choice = input("Enter your choice: ")

    # input of details of account holder
    
    if choice=="1":
        while True:
            acc_holder_name=input("enter your name with which you want to open an account:\n").lower().strip()
            if acc_holder_name:
                break
            print("name can not be empty")
            
        print("Name:",acc_holder_name)
        while True:
            acc_type=input("enter your choice of account you want to open saving or current:\n")
            if acc_type not in ("saving","current"):
                print("enter the valid input")
            else:
                break
        print("Account Type:",acc_type)
        while True:
            try:
                initial_balance=float(input("enter the initial amount you want to deposit:\n"))
                if initial_balance<0:
                    print("balance can not be negative")
                else:
                    break
            except ValueError:
                print("please enter numeric value")
        print("Initial Balance:",initial_balance)
    
        # account number generation 
        account_number=account_number+1
        print("Your account number is:\n",account_number)
        
        # storing of account details in dictionary
        account_details={
        "name":acc_holder_name,
        "type":acc_type,
        "balance":initial_balance,
        }
        accounts[account_number]=account_details
        print("account created succesfully")
        print("account number is:",account_number)
        print("account details are:",accounts[account_number])
        transactions.append({
    "acc_no": account_number,
    "type": "account_opened",
    "amount": initial_balance
})

        
    # view account details
    elif choice=="2":
        acc_no=int(input("enter the valid account number to view the details of customer"))
        if acc_no in accounts:
            print("-------Account Details-------")
            print("Account Holder:",accounts[acc_no]["name"])
            print("Account Type:",accounts[acc_no]["type"])
            print("Balance:",accounts[acc_no]["balance"])
        else:
            print("account not found")
        
    # transactions
    # deposits
    elif choice=="3":
        acc_no=int(input("enter your account number in which you want to deposit your amount:\n"))
        if acc_no in accounts:
            while True:
                try:
                    deposit_amount=float(input("enter the amount your want to deposit in your account:\n"))
                    if deposit_amount>0:
                        accounts[acc_no]["balance"]=accounts[acc_no]["balance"]+deposit_amount
                        transactions.append({
    "acc_no": acc_no,
    "type": "deposit",
    "amount": deposit_amount
})

                        print("your updated balance is:\n",accounts[acc_no]["balance"])
                        break
                    else:
                        print("amount must be greater than zero")
                    
                except ValueError:
                    print("please enter numeric value")
        else:
            print("account not found")
    # withdraw
    elif choice=="4":
        acc_no=int(input("enter your account number in which you want to with draw from your amount:\n"))
        if acc_no in accounts:
            while True:
                try:
                    withdraw_amount=float(input("enter the amount your want to withdraw from your account:\n"))
                    if withdraw_amount<=0:
                        print("amount must be greater than zero")
                    elif withdraw_amount>accounts[acc_no]["balance"]:
                        print("insufficient balance")
                    else:
                        accounts[acc_no]["balance"]=accounts[acc_no]["balance"]-withdraw_amount
                        transactions.append({
    "acc_no": acc_no,
    "type": "withdraw_amount",
    "amount": withdraw_amount
})

                        print("your updated balance is:\n",accounts[acc_no]["balance"])
                        break
                    
                except ValueError:
                    print("please enter numeric value")
        else:
            print("account not found")
    
    # transfer
    elif choice=="5":
        sender_acc=int(input("enter account number of sender:\n"))
        reciever_acc=int(input("enter account number of reciever:\n"))
        if sender_acc in accounts and reciever_acc in accounts:
            while True:
                try:
                    transfer_amount=float(input("enter the amount to tranfer:\n"))
                    if transfer_amount<=0:
                        print("enter amount greater than zero")
                    elif transfer_amount>accounts[sender_acc]["balance"]:
                        print("insufficient balance in sender amount")
                    else:
                        accounts[sender_acc]["balance"]-=transfer_amount
                        accounts[reciever_acc]["balance"]+=transfer_amount
                        transactions.append({
    "acc_no": sender_acc,
    "type": "transfer_sent",
    "amount": transfer_amount
})

                        transactions.append({
    "acc_no": reciever_acc,
    "type": "transfer_received",
      "amount": transfer_amount
    })

                        print("transfer sucessful")
                        print("sender remaining balance:",accounts[sender_acc]["balance"])
                        print("receiver remaining balance:",accounts[reciever_acc]["balance"])
                        break
                except ValueError:
                    print("please enter numeric value")
        else:
            print("sender or reciever account not found")
    elif choice == "6":
        acc_no = int(input("Enter account number:\n"))
        found = False
    
        print("\n----------------------------------------")
        print("Account Number :", acc_no)
        print("----------------------------------------")
        print(f"{'Type':20} {'Amount'}")
        print("----------------------------------------")
    
        for t in transactions:
            if t["acc_no"] == acc_no:
                print(f"{t['type']:20} {t['amount']}")
                found = True
    
        if not found:
            print("No transactions found for this account")
    
        print("----------------------------------------")

    elif choice=="7":
        with open("accounts.pkl", "wb") as f:
            pickle.dump(accounts, f)
        
        with open("meta.pkl", "wb") as f:
            pickle.dump(account_number, f)

        with open("transactions.pkl", "wb") as f:
            pickle.dump(transactions, f)

        print("data saved successfully")
        break
    
        
    else:
        print("invalid choice try again")


print("Thank you! for your time")                 
                
            
```
